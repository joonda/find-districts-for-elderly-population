```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rc
import warnings
import os

if os.name == 'nt':
    print("This is a Windows environment.")
    rc('font', family='Malgun Gothic')
elif os.name == 'posix':
    print("This is a Mac or Linux environment.")
    rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False

warnings.filterwarnings("ignore", category=UserWarning)
```

```python
raw_pop = pd.read_excel('../data/자치구별+고령인구(추계인구)_20241019165325.xlsx')
raw_pop.head()
```

```python
pop = raw_pop.drop(index=[0, 1, 2], columns=['자치구별(1)'])
pop.columns = ["구별", "노령인구", "노령인구_남자", "노령인구_여자"]
pop.reset_index(drop=True, inplace=True)
```

```python
raw_market = pd.read_csv('../data/소상공인시장진흥공단_전통시장현황_20240719.csv', encoding='cp949')
raw_market.head()
```

```python
market = raw_market[raw_market['시도'] == '서울특별시']
```

```python
market = market['시군구'].value_counts().reset_index().rename(columns={'count' : '시장개수', '시군구' : '구별'})
market.head()
```

* https://data.seoul.go.kr/dataList/10052/S/2/datasetView.do

```python
raw_park = pd.read_excel('../data/공원_20241218013757.xlsx')
raw_park.head()
```

```python
park = raw_park[['자치구별(2)', '공원수 (개소)']]
park.rename(columns={'자치구별(2)' : '구별', '공원수 (개소)' : '공원수'}, inplace=True)
park.drop([0,26], inplace=True)
park = park.reset_index(drop=True)
park.head()
```

```python
culture = park.merge(right=market, right_on='구별', left_on='구별')
culture.head()
```

```python
result_culture = culture.merge(right=pop, right_on='구별', left_on='구별')
result_culture.head()
```

```python
result_culture = result_culture.drop(columns=['노령인구_남자', '노령인구_여자'])
result_culture.head()
```

```python
result_culture['문화시설개수'] = result_culture['공원수'] + result_culture['시장개수']
result_culture['문화시설개수(10K)'] = result_culture['문화시설개수'] / result_culture['노령인구'] * 10000
result_culture.head()
```

```python
result_culture.to_csv('../data/output/culture_f.csv', index=False)
```
