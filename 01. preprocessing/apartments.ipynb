```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc
import warnings
import os

if os.name == 'nt':
    print("This is a Windows environment.")
    rc('font', family='Malgun Gothic')
elif os.name == 'posix':
    print("This is a Mac or Linux environment.")
    rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False

warnings.filterwarnings("ignore", category=UserWarning)
```

```python
raw_apart = pd.read_excel('../data/아파트(전월세)_실거래가_20241219203300.xlsx')
raw_officetel = pd.read_excel('../data/오피스텔(전월세)_실거래가_20241219203419.xlsx')

raw_apart.shape, raw_officetel.shape
```

```python
raw_apart.head()
```

```python
apart = raw_apart[raw_apart['전용면적(㎡)'] <= 60]
tel = raw_officetel[raw_officetel['전용면적(㎡)'] <= 60]

apart.shape, tel.shape
```

```python
apart = apart[['시군구', '단지명', '전월세구분', '전용면적(㎡)', '보증금(만원)', '월세금(만원)']]
tel = tel[['시군구', '단지명', '전월세구분', '전용면적(㎡)', '보증금(만원)', '월세금(만원)']]

apart.reset_index(inplace=True, drop=True)
tel.reset_index(inplace=True, drop=True)

apart.shape, tel.shape
```

```python
si_list = []
gu_list = []

for i, v in apart.iterrows():
    address = v['시군구']
    si = address.split()[0]
    gu = address.split()[1]
    si_list.append(si)
    gu_list.append(gu)
    
len(si_list), len(gu_list)
```

```python
apart['구별'] = gu_list
apart['구별'].value_counts()
```

```python
si_list = []
gu_list = []

for i, v in tel.iterrows():
    address = v['시군구']
    si = address.split()[0]
    gu = address.split()[1]
    si_list.append(si)
    gu_list.append(gu)
    
len(si_list), len(gu_list)
```

```python
tel['구별'] = gu_list
tel['구별'].value_counts()
```

```python
tel.head()
```

```python
apart.head()
```

```python
apart.shape
```

```python
cost_list = []

for idx, val in apart.iterrows():
    if val['전월세구분'] == '전세':
        cost = (val['보증금(만원)'] / 12) * 0.05
        cost_list.append(cost)
    elif val['전월세구분'] == '월세':
        cost = val['월세금(만원)'] + ((val['보증금(만원)'] / 12) * 0.05)
        cost_list.append(cost)
        
len(cost_list)
```

```python
apart['환산월세금(만원)'] = cost_list

apart.head()
```

```python
apart.groupby(['구별'])['환산월세금(만원)'].mean().reset_index()
```

```python
cost_list = []

for idx, val in tel.iterrows():
    if val['전월세구분'] == '전세':
        cost = (val['보증금(만원)'] / 12) * 0.05
        cost_list.append(cost)
    elif val['전월세구분'] == '월세':
        cost = val['월세금(만원)'] + ((val['보증금(만원)'] / 12) * 0.05)
        cost_list.append(cost)
        
len(cost_list)
```

```python
tel['환산월세금(만원)'] = cost_list
tel.head()
```

```python
result_tel = tel.groupby(['구별'])['환산월세금(만원)'].mean().reset_index()
result_tel.head()
```

```python
result_tel.to_csv('../data/output/apartments_f.csv', index=False)
```
